# Creating dataframes

```python
##specify values for each column.
df1=pd.DataFrame({'a':[1,2,3],
                 'b':[3,4,0],
                 'c':[44,43,6]})# output: a dataframe with columns a,b,c and index:RangeIndex(start=0, stop=3, step=1)

df2=pd.DataFrame({'a':[1,1,1],
                 'b':[2,2,2],
                 'c':[3,3,3]},
                 index=['first','second','thired'])

#specify values for each row
df3=pd.DataFrame([['row1','row1','row1'],
                  ['row2','row2','row2'],
                  ['row3','row3','row3']],
                 index=['first','second','thired'],
                 columns=['a','b','c'])

#create dataframe with a multiIndex
df4=pd.DataFrame({'a':['col1','col1','col1'],
                  'b':['col2','col2','col2'],
                  'c':['col3','col3','col3']},
                 index=pd.MultiIndex.from_tuples([('a',0),('a',2),('b',2)],names=['first index','second index']))



df4
```

# Reshaping data

```python
#pd.melt(df): gather columns into rows
print('df1\n',df1)
print('pd.melt(df1)\n',pd.melt(df1))
```

```python
#pd.concat([df1,df2]): append rows of dataframe

df1=pd.DataFrame([['a',1],['b',2]],columns=['letter','number'])

df2=pd.DataFrame([['c',3],['d',4]],columns=['letter','number'])

print('df1\n',df1)
print('df2\n',df2)
print('pd.concat([df1,df2])\n',pd.concat([df1,df2]))
```

```python
#pd.concat([df1,df2],axis=1): append columns of dataframes
df4 = pd.DataFrame([['bird', 'polly'], ['monkey', 'george']],
                   columns=['animal', 'name'])
pd.concat([df1, df4], axis=1)
```

```python
#df.pivot(columns='var',values='val'): spread rows into columns

df = pd.DataFrame({'foo': ['one', 'one', 'one', 'two', 'two',
                           'two'],
                   'bar': ['A', 'B', 'C', 'A', 'B', 'C'],
                   'baz': [1, 2, 3, 4, 5, 6],
                   'zoo': ['x', 'y', 'z', 'q', 'w', 't']})
#df
#df.pivot(index='foo',columns='bar',values='baz')
#df.pivot(index='foo', columns='bar')['baz']
df.pivot(index='foo', columns='bar', values=['baz', 'zoo'])
```

```python
#df.sort_values('mpg'): order rows by values of a column(low to high)
df=pd.DataFrame({'col1': ['A', 'A', 'B', pd.NA, 'D', 'C'],
    'col2': [2, 1, 9, 8, 7, 4],
    'col3': [0, 1, 9, 4, 2, 3],
    'col4': ['a', 'B', 'c', 'D', 'e', 'F']})
print('initial df\n',df)
print('sorted\n',df.sort_values(by=['col1']))
#check if this function is inplace?
print('initial df\n',df)
```

```python
#df.rename(columns={'y':'year'}):rename the columns of a Dataframe
#df.rename(index={0: "x", 1: "y", 2: "z"}):rename the indices of a dataframe
df = pd.DataFrame({"y": [2001, 2002, 2003], "sal": [4, 5, 6]})
print('original df\n',df)
df.rename(columns={'y':'year','sal':'salary'},index={0:'first'})
print('changed df\n',df)
```

```python
#df.sort_index(): Sort the index of a DataFrame

df=pd.DataFrame({'a':[1,1,1],
                 'b':[2,2,2],
                 'c':[3,3,3]},
                 index=[29,0,1])
df.sort_index(key=lambda x: x*(-1))

#A key function can be specified which is applied to the index before sorting. For a MultiIndex this is applied to each level separately.
```

```python
#df.reset_index(): Reset index of DataFrame to row numbers, moving index to columns.
df = pd.DataFrame([('bird', 389.0),
                   ('bird', 24.0),
                   ('mammal', 80.5),
                   ('mammal', None)],
                  index=['falcon', 'parrot', 'lion', 'monkey'],
                  columns=('class', 'max_speed'))
df.reset_index()
```

# Subset Variables - columns

# Subsets - rows and columns

```python
#Use df.loc[] and df.iloc[] to select only rows, only columns or both

# Create a simple dataframe
data = {'Name': ['John', 'Anna', 'Peter', 'Linda'],
        'Age': [28, 24, 35, 32],
        'City': ['New York', 'Paris', 'Berlin', 'London']}
df=pd.DataFrame(data)
df
```

```python
print("\nSelecting a single row using df.loc[]:")
print(df.loc[1])
```

```python
print("\nSelecting multiple rows using df.loc[]:")
print(df.loc[[1,2]])
```

```python
print("\nSelecting rows and columns using df.loc[]:")
print(df.loc[1:3, 'Name':'Age'])
```

```python
print("\nSelecting a single row using df.iloc[]:")
print(df.iloc[1])
```

```python
print("\nSelecting rows and columns using df.iloc[]:")
print(df.iloc[1:3, 0:2])
```

```python
# Select all rows and some columns by label
selected_columns = df.loc[:, ['Name', 'Age']]
print(selected_columns)
```

```python
# Select all rows and some columns by index
selected_columns=df.iloc[:,[0,1]]
print(selected_columns)
```

```python
#Use df.at[] and df.iat[] to access a single value by row and column First index selects rows, second index columns.
print("\nAccessing a single value using df.at[]:")
print(df.at[1,'Name'])
```

```python
print("\nAccessing a single value using df.iat[]:")
print(df.iat[1, 0])
```


# Using query

query() allows Boolean expressions for filtering
rows

```python
older_than_30 = df.query('Age > 30')
print(older_than_30)
```

In this example, df.query('Age > 30') returns a new DataFrame containing only the rows where the Age column is greater than 30.

The string you pass to query() can be any expression that evaluates to a boolean value. You can use and, or, and not operators, and you can compare columns to other columns or to constants.

```python
df.query('City.str.startswith("B")')
```

# Make New Columns

```python
#DataFrame.assign(**kwargs): Assign new columns to a DataFrame
df = pd.DataFrame({'temp_c': [17.0, 25.0]},
                  index=['Portland', 'Berkeley'])
df=df.assign(temp_f = lambda x: x['temp_c'] * 9/5+32, humidity='low')
```

```python
df
```

```python
#df['Volume'] = df.Length*df.Height*df.Depth : Add single column

df['chertopert']=df.temp_c *df.temp_f
df
```

```python
#pd.qcut(df.col, n, labels=False): Bin column into n buckets.

pd.qcut(df.temp_c, 10, labels=False)
```
